```
import numpy as np
import pandas as pd
import copy
import matplotlib.pyplot as plt
import datetime
import time
%matplotlib inline
```

```
# read cleanned training set for Exploratory Data Analysis
df = pd.read_csv('newTrain.csv')
```

```
# EDA for each variable. These plots help us select informative features. 
# We let our classifier do feature selection for us.
import seaborn as sns

plt.figure(figsize=(12,30*4))
import matplotlib.gridspec as gridspec
features = df.iloc[:,0:19].columns
gs = gridspec.GridSpec(30, 1)
for i, feature in enumerate(df[features]):
    ax = plt.subplot(gs[i])
    sns.distplot(df[feature][df.Def_ind == 1], bins=50)
    sns.distplot(df[feature][df.Def_ind == 0], bins=50)
    ax.set_xlabel('')
    ax.set_title('Feature: ' + str(feature))
plt.show()
```
